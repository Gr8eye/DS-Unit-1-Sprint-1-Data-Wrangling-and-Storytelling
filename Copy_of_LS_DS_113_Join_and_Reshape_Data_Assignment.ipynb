```python
#imports first
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# After concluding that order_products and product names would be the only frames I needed I merged them
products = pd.read_csv('products.csv')
products
```

```python
product_orders = pd.merge(order_products, products, on = 'product_id', how= 'inner')
```

```python
product_orders
```

```python
# I chose the columns that I would need
cols = ['order_id', 'product_id', 'product_name']
```

```python
# and applied the condition to the new dataframe
POcount = product_orders[cols]
POcount
```

```python
# From this I decided to create a new dataframe (end goal)

Count = pd.value_counts(POcount['product_name']).to_frame().reset_index()
Count
```

```python
#Taking the *Replicate with a grain of salt I opted to focus on Table 2
import pandas as pd
import numpy as np

table1 = pd.DataFrame(
    [[np.nan, 2],
     [16,    11], 
     [3,      1]],
    index=['John Smith', 'Jane Doe', 'Mary Johnson'], 
    columns=['treatmenta', 'treatmentb'])

table2 = table1.T

table2
```

```python
# used melt to give column names
T2Tidy = table2.reset_index().melt(id_vars='index').rename(columns= {
    'variable': 'Name', 
    'index': 'Treatment', 
    'value': 'Result'
})
T2Tidy
```

```python
#Table2 to tidy
#Cleaned up the columns
T2Tidy['Treatment'] = T2Tidy['Treatment'].str.replace('treatment', '')
T2Tidy['Treatment'] = T2Tidy['Treatment'].str.replace('a', 'A')
T2Tidy['Treatment'] = T2Tidy['Treatment'].str.replace('b','B')
T2Tidy
```

```python
#Tidy to Table 2
T2Tidy.pivot_table(index='Name', columns='Treatment', values='Result')
```
